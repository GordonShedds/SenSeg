## In this tutorial we will use the same multiclass data from previous one and train model on it

```python
from utils.featurizer import SentinelFeaturizer
from utils.trainer import train_model
from utils.predictor import predict
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
featurizer = SentinelFeaturizer(image='sentinel_data_4_bands.tif',
                                mask='mask.geojson',
                                mask_label_column='class')
```

## We will not normalize data, as random forest algorithm doesnt actually need it

```python
X, y = featurizer.get_features(window_size=3)
```

```python
model = train_model(X, y)
```

## Trained model can be saved and loaded via joblib: dump/load

## Now lets make prediction on new data:

- first, we read new image using SentinelFeaturizer
- next, we extract X features from it
- last, we use predict on this features

```python
prediction = predict(image='new_image_for_prediction.tif',
                     model=model,
                     window_size=3)

plt.imshow(prediction)
```

## We can also perform simple postprocessing as opening + closing operations:

It will clear some tiny noise and fill small holes

WARNING: this feature currently work correctly for binary classification task, WIP

```python
prediction = predict(image='new_image_for_prediction.tif',
                     model=model,
                     window_size=3,
                     opening_size=2)

plt.imshow(prediction)
```

## Saving results as georeferenced image:

Just pass save_path argument in predict function

```python
prediction = predict(image='new_image_for_prediction.tif',
                     model=model,
                     window_size=3,
                     save_path='prediction.tif')
```
